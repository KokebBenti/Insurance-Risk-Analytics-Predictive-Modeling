```python
#encode test data
from scipy import sparse
cat_cols2 = df_te.select_dtypes(include=['object', 'category','bool']).columns
num_df = df_te.drop(columns=cat_cols2)
X_num = sparse.csr_matrix(num_df.values)
ohe_array = ohe.transform(df_te[cat_cols2])
ohe_cat_cols = ohe.get_feature_names_out(cat_cols2).tolist()
X_all = sparse.hstack([X_num,ohe_array])
all_cols=list(num_df.columns)+ohe_cat_cols
df_te = pd.DataFrame.sparse.from_spmatrix(X_all, columns=all_cols)
```

```python
x,y=df_tr.drop(['TotalClaims'], axis=1), df_tr['TotalClaims']
x_te,y_te=df_te.drop(['TotalClaims'], axis=1), df_te['TotalClaims']
```

```python
x.info()
```

```python
# Linear regression
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(x, y)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_lin= linreg.predict(x)
mse=mean_squared_error(y,yp_lin)
mae=mean_absolute_error(y,yp_lin)
r2=r2_score(y, yp_lin)
print(mse,mae,r2)
```

```python
#error on test set
yp_t= linreg.predict(x_te)
mse=mean_squared_error(y_te,yp_t)
maet1=mean_absolute_error(y_te,yp_t)
r2=r2_score(y_te,yp_t)
print(mse,mae,r2)
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor( max_depth=5, min_samples_split=10, min_samples_leaf=5,random_state=42)
tree_reg.fit(x, y)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_dt= tree_reg.predict(x)
mse=mean_squared_error(y,yp_dt)
mae3=mean_absolute_error(y,yp_dt)
r2=r2_score(y, yp_dt)
print(mse,mae3,r2)
```

```python
#error on test set
yp_t2= tree_reg.predict(x_te)
mse=mean_squared_error(y_te,yp_t2)
maet3=mean_absolute_error(y_te,yp_t2)
r2=r2_score(y_te,yp_t2)
print(mse,mae,r2)
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rfr_model = RandomForestRegressor(n_estimators=20,max_depth=10, min_samples_leaf=5,random_state=42,n_jobs=-1)
rfr_model.fit(x, y)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_rfr= rfr_model.predict(x)
mse=mean_squared_error(y,yp_rfr)
mae4=mean_absolute_error(y,yp_rfr)
r2=r2_score(y, yp_rfr)
print(mse,mae4,r2)
```

```python
#error on test set
yp_t3= rfr_model.predict(x_te)
mse=mean_squared_error(y_te,yp_t3)
maet3=mean_absolute_error(y_te,yp_t3)
r2=r2_score(y_te,yp_t3)
print(mse,mae,r2)
```

```python
#XGBoost
import xgboost
xgb_reg = xgboost.XGBRegressor(n_estimators=1000, learning_rate=0.06, max_depth=4, random_state=42,n_jobs=-1)
xgb_reg.fit(x, y)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_xgb= xgb_reg.predict(x)
mse=mean_squared_error(y,yp_xgb)
mae5=mean_absolute_error(y,yp_xgb)
r2=r2_score(y, yp_xgb)
print(mse,mae5,r2)
```

```python
#error on test set
yp_t4= xgb_reg.predict(x_te)
mse=mean_squared_error(y_te,yp_t4)
maet5=mean_absolute_error(y_te,yp_t4)
r2=r2_score(y_te,yp_t4)
print(mse,maet5,r2)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [100, 300],'max_depth': [3, 5, 7], 'learning_rate': [0.01, 0.1, 0.2]}
grid_search = GridSearchCV(estimator=xgb_reg,  param_grid=param_grid,  cv=2, scoring='neg_mean_squared_error',verbose=1, n_jobs=-1)
grid_search.fit(x, y)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x)
print("Best Parameters:", grid_search.best_params_)
```

```python
mse=mean_squared_error(y,y_pred)
mae=mean_absolute_error(y,y_pred)
r2=r2_score(y,y_pred)
print(mse,mae,r2)
```
