```python
#Split data into training and testing
df_tr,df_te=Data_Cleaning.test_data(df)
x_tr,y_tr=df_tr.drop(['TotalClaims','TransactionMonth'], axis=1), df_tr['TotalClaims']
x_te,y_te=df_te.drop(['TotalClaims','TransactionMonth'], axis=1), df_te['TotalClaims']
```

```python
# Linear regression
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(x_tr, y_tr)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_lin= linreg.predict(x_tr)
mse=mean_squared_error(y_tr,yp_lin)
mae=mean_absolute_error(y_tr,yp_lin)
r2=r2_score(y_tr, yp_lin)
print(mse,mae,r2)
```

```python
#error on test set
yp_t= linreg.predict(x_te)
mse=mean_squared_error(y_te,yp_t)
mae=mean_absolute_error(y_te,yp_t)
r2=r2_score(y_te,yp_t)
print(mse,mae,r2)
```

```python
# SGD regression
from sklearn.linear_model import SGDRegressor
sgdreg=SGDRegressor(alpha=1e-5, learning_rate='constant',  eta0=0.01,  max_iter=10000, tol=1e-3,  random_state=42)
sgdreg.fit(x_tr, y_tr)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_sgd= sgdreg.predict(x_tr)
mse=mean_squared_error(y_tr,yp_sgd)
mae=mean_absolute_error(y_tr,yp_sgd)
r2=r2_score(y_tr, yp_sgd)
print(mse,mae,r2)
```

```python
#error on test set
yp_t2= sgdreg.predict(x_te)
mse=mean_squared_error(y_te,yp_t2)
mae=mean_absolute_error(y_te,yp_t2)
r2=r2_score(y_te,yp_t2)
print(mse,mae,r2)
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor( max_depth=5, min_samples_split=10, min_samples_leaf=5,random_state=42)
tree_reg.fit(x_tr, y_tr)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_dt= tree_reg.predict(x_tr)
mse=mean_squared_error(y_tr,yp_dt)
mae=mean_absolute_error(y_tr,yp_dt)
r2=r2_score(y_tr, yp_dt)
print(mse,mae,r2)
```

```python
#error on test set
yp_t3= tree_reg.predict(x_te)
mse=mean_squared_error(y_te,yp_t)
mae=mean_absolute_error(y_te,yp_t)
r2=r2_score(y_te,yp_t3)
print(mse,mae,r2)
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rfr_model = RandomForestRegressor(n_estimators=20,max_depth=10, min_samples_leaf=5,random_state=42,n_jobs=-1)
rfr_model.fit(x_tr, y_tr)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_rfr= rfr_model.predict(x_tr)
mse=mean_squared_error(y_tr,yp_rfr)
mae=mean_absolute_error(y_tr,yp_rfr)
r2=r2_score(y_tr, yp_rfr)
print(mse,mae,r2)
```

```python
#XGBoost
import xgboost
xgb_reg = xgboost.XGBRegressor(n_estimators=1000, learning_rate=0.06, max_depth=4, random_state=42,n_jobs=-1)
xgb_reg.fit(x_tr, y_tr)
```

```python
#error on training set
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
yp_xgb= xgb_reg.predict(x_tr)
mse=mean_squared_error(y_tr,yp_xgb)
mae=mean_absolute_error(y_tr,yp_xgb)
r2=r2_score(y_tr, yp_xgb)
print(mse,mae,r2)
```

```python
#error on test set
yp_t4= xgb_reg.predict(x_te)
mse=mean_squared_error(y_te,yp_t4)
mae=mean_absolute_error(y_te,yp_t4)
r2=r2_score(y_te,yp_t4)
print(mse,mae,r2)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 'n_estimators': [100, 300],    'max_depth': [3, 5, 7],    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 1.0],    'colsample_bytree': [0.7, 1.0]}
grid_search = GridSearchCV(estimator=xgb_reg,  param_grid=param_grid,  cv=3, scoring='neg_mean_squared_error',
    verbose=1,    n_jobs=-1)

grid_search.fit(x_tr, y_tr)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_tr)
print("Best Parameters:", grid_search.best_params_)
```

```python
mse=mean_squared_error(y_tr,y_pred)
mae=mean_absolute_error(y_tr,y_pred)
r2=r2_score(y_tr,y_pred)
print(mse,mae,r2)
```
