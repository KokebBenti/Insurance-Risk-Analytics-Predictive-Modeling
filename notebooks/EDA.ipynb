Bivariate or Multivariate Analysis

Monthly change

```python
df['month'] = df['TransactionMonth'].dt.to_period('M')
g=df.groupby(['month', 'Province'])['TotalClaims'].sum().unstack()
g2=df.groupby(['month', 'Province'])['TotalPremium'].sum().unstack()
```

```python
g
```

```python
g.plot(kind='bar',figsize=(14, 12))
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('TotalClaims')
plt.title('Total Claims')
plt.show()
```

```python
g2.plot(kind='bar',figsize=(14, 12))
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('TotalPremium')
plt.title('Total Premium')
plt.show()
```

```python
#correlation between total claims and total premiums
df['TotalClaims'].corr(df['TotalPremium'])
```
